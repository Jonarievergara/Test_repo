```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df_all = pd.read_csv("df_all2.csv", encoding='latin1')
df_all
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_unique_school = df_all.drop_duplicates('School ID', keep='first').copy()
df_unique_school
```

```python
df_unique_school["total_teachers"] = df_unique_school['teachers.instructor'] + df_unique_school['teachers.mobile'] + df_unique_school['teachers.regular'] + df_unique_school['teachers.sped']
```

```python
df_unique_school['st_ratio'] = df_unique_school['Enrolment'] /df_unique_school['total_teachers']
```

```python
df_unique_school['st_ratio'] = df_unique_school['st_ratio'].replace(np.inf, np.nan)
df_unique_school['st_ratio'] = df_unique_school['st_ratio'].replace(-np.inf, np.nan)
df_unique_school['st_ratio']
```

```python
sns.histplot(data=df_unique_school, x="st_ratio")
```

```python
sns.set_theme(style="whitegrid")
ax = sns.boxplot(x=df_unique_school["st_ratio"])
```

```python
df_unique_school = df_unique_school[df_unique_school['st_ratio'].notna()]
df_unique_school
```

```python
Q1 = df_unique_school['st_ratio'].quantile(0.25)
Q3 = df_unique_school['st_ratio'].quantile(0.75)
IQR = Q3 - Q1
df_no_outliers = (df_unique_school[(df_unique_school['st_ratio'] >= Q1 - (1.5 * IQR)) & 
                                   (df_unique_school['st_ratio'] <= Q3 + (1.5 * IQR))])
```

```python
sns.histplot(data=df_no_outliers, x="st_ratio", palette = 'husl')
```

```python
ax = sns.boxplot(x=df_no_outliers["st_ratio"], palette = 'husl')
```

```python
stats.normaltest(df_no_outliers["st_ratio"])
```

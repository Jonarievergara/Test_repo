```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_line = pd.read_excel("DataViz_Exercises - Data_FiNAL.xlsx", sheet_name="Number 2")
df_line = df_line.iloc[0:11]
df_line
```

```python
df_bar = pd.read_excel("DataViz_Exercises - Data_FiNAL.xlsx", sheet_name="Number 4")
df_bar = df_bar.iloc[0:7]
df_bar
```

```python
cols = df_line.columns
cols
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data = df_line, x="Year", y=cols[1], color="red")
sns.lineplot(data = df_line, x="Year", y=cols[2], color="green")
ax2 = plt.twinx()
sns.lineplot(data = df_line, x="Year", y=cols[3], ax=ax2, linewidth=5)
ax2.set_ylim(0,100)
plt.title("Trend of Hours used of Mobile vs Desktop/Laptop over Percent of Social Media Usage", fontsize=20)
```

```python
data = pd.melt(df_bar, id_vars=' ')
data
```

```python
plt.figure(figsize=(20,8))
g = sns.barplot(data=data, x=' ', y='value', hue="variable")
plt.ylabel("No of users", fontsize=15)
plt.xlabel("Age range", fontsize=15)
plt.title("Social Media Users in the Philippines vs Worldwide", fontsize=25)
```
